```python
countries = ['United States', 'Finland']

US_GDP = US_Score.iloc[0, 6]
US_SS = US_Score.iloc[0, 7]
US_HLE = US_Score.iloc[0, 8]
US_FLC = US_Score.iloc[0, 9]
US_Gen = US_Score.iloc[0, 10]
US_Corr = US_Score.iloc[0, 11]

Fin_GDP = Fin_Score.iloc[0, 6]
Fin_SS = Fin_Score.iloc[0, 7]
Fin_HLE = Fin_Score.iloc[0, 8]
Fin_FLC = Fin_Score.iloc[0, 9]
Fin_Gen = Fin_Score.iloc[0, 10]
Fin_Corr = Fin_Score.iloc[0, 11]

GDP = [US_GDP, Fin_GDP]
SS = [US_SS, Fin_SS]
HLE = [US_HLE, Fin_HLE]
FLC = [US_FLC, Fin_FLC]
Gen = [US_Gen, Fin_Gen]
Corr = [US_Corr, Fin_Corr]

x = range(len(countries))

plt.bar(x, GDP, label = 'Explained by GDP per Capita')
plt.bar(x, SS, bottom = GDP, label = 'Explained by Social Support')
plt.bar(x, HLE, bottom=[i + j for i, j in zip(GDP, SS)], label='Explained by Healthy Life Expectancy')
plt.bar(x, FLC, bottom=[i + j for i, j in zip(GDP, SS, HLE)], label='Explained by Freedom to make Life Choices')

plt.xlabel('Country')
plt.ylabel('Variable Values')
plt.title('Happiness Score of the US vs Finland')
plt.xticks(x, countries)
plt.legend()
```

```python
US = df1[df1['Country name'] == 'United States']
Fin = df1[df1['Country name'] == 'Finland']
```

```python
x1 = US['year']
x2 = Fin['year']
y1_Gen = US['Generosity']
y2_Gen = Fin['Generosity']

plt.plot(x1, y1_Gen, label='US', color='blue', marker='o')
plt.plot(x2, y2_Gen, label='Finland', color='red', marker='s')

plt.xlabel('Years')
plt.ylabel('Generosity')
plt.title('Generosity in US vs. Finland')
plt.legend()

plt.show
```
